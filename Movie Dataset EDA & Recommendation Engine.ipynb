# Missing Values Imputation

```python
# The function we defined to calculate the number and percentage of null values and present them in a DataFrame.

def analyze_missing_values(data):
    missing_df = pd.DataFrame({
        "count":
        data.isnull().sum(),
        "percentage":
        round(data.isnull().mean() * 100, 2)
    })
    missing_df = missing_df[missing_df['count'] != 0].sort_values(
        by='count', ascending=True)
    return missing_df
```

```python
analyze_missing_values(data)
```

```python
# Due to the high number of NaN values in the 'gross' and 'budget' columns, we are dropping all rows that have NaN values in these columns.

# For this, we are using the NumPy 'isnan' function with a negation (~) to drop all rows with NaN values in these columns.
data = data[~np.isnan(data['gross']) & ~np.isnan(data['budget'])]

analyze_missing_values(data)
```

```python
# It looks at the total number of Null values in each row, keeping only the rows where the total sum of Null values is less than two.

data = data[data.isnull().sum(axis=1) <= 2]

analyze_missing_values(data)
```

```python
# Let's fill in the missing values.

# We are using the mean value for numerical columns.
data['num_critic_for_reviews'].fillna(data['num_critic_for_reviews'].mean(),
                                      inplace=True)
data['duration'].fillna(data['duration'].mean(), inplace=True)

# We are using the mode for categorical columns.
data['language'].fillna(data['language'].mode()[0], inplace=True)

# For categorical data that cannot be filled with statistical values like actor names, we are using the expression "Unknown Actor".
data['actor_2_name'].fillna('Unknown Actor', inplace=True)
data['actor_3_name'].fillna('Unknown Actor', inplace=True)

analyze_missing_values(data)
```

**mlxtend (Machine Learning Extensions):**

- mlxtend is a Python library that provides useful tools and algorithms for machine learning, data science, and data mining applications.

- mlxtend.preprocessing is the preprocessing module of the mlxtend library, which contains various classes and functions for data preprocessing operations.

- This module facilitates a series of preprocessing steps necessary to prepare datasets for machine learning algorithms.

- Among the operations are feature scaling, converting categorical data to numerical data, and converting operations in datasets to binary formats.





**TransactionEncoder:**
- TransactionEncoder is a class used to convert items within a sequence of transactions (e.g., shopping basket transactions or, as in this case, a list of movie genres) into binary format.

- It creates a binary matrix (True/False or 1/0) indicating whether each item is present in each transaction. This facilitates determining the presence of items for analyses such as frequent itemset mining.

- In this particular case, TransactionEncoder takes lists containing movie genres and generates a binary matrix indicating whether each genre exists in the movies.

- As a result, each column represents a movie genre, and each row indicates whether that genre is present in the respective movie (1 or 0).

- This transformation prepares the dataset for analysis and modeling of genres.

https://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/




```python
data['genres']
```

```python
# Split the genres in the 'genres' column by the '|' character and convert them into a list.

x = data['genres'].str.split('|')
```

```python
# Create a TransactionEncoder object.

te = TransactionEncoder()
```

```python
# Apply the fit and transform operations to the created list using TransactionEncoder.

x = te.fit_transform(x)
```

```python
# Convert the result to a pandas DataFrame and use column names from TransactionEncoder.

x = pd.DataFrame(x, columns=te.columns_)
```

```python
x.head()
```

```python
# Let's convert the data from boolean to integer so that we can perform calculations.

genres = x.astype('int')

genres.head()
```

```python
# Now, let's add movie titles to the first column for better understanding of the data.

genres.insert(0, 'movie_title', data['movie_title'])

genres.head()
```

```python
# Let's make these movie titles the index of the data.

genres = genres.set_index('movie_title')

genres.head()
```

```python
# Let's build a recommendation engine to find similar genres.

# Let's calculate the correlation between the target genre and all other genres.


def recommendation_genres(gen):
    gen = genres[gen]
    similar_genres = genres.corrwith(gen)
    similar_genres = similar_genres.sort_values(ascending=False)
    similar_genres = similar_genres.iloc[1:]
    return similar_genres.head(3)
```

**corrwith Function**

- It is used for DataFrame objects in the Pandas library and computes column-wise correlations between a DataFrame and another Series or DataFrame.

- This function uses the Pearson correlation coefficient to measure the relationship between two datasets.

- Here, it calculates the correlation between the target genre and all other genres.

- It is an effective method for identifying similar genres because high correlation values indicate a strong relationship between genres. This way, users can discover new genres that match their interests.

```python
recommendation_genres('Comedy')
```

# Recommending similar Movies

```python
# We are creating a new structure by transposing the 'genres' DataFrame.

x = genres.transpose()
x.head()
```

```python
# A function to find similar movies for a given movie.


def recommendation_movie(movie):
    movie = x[movie + '\xa0']
    similar_movies = x.corrwith(movie)
    similar_movies = similar_movies.sort_values(ascending=False)
    similar_movies = similar_movies.iloc[1:]
    return similar_movies.head(20)
```

**Non-breaking Space and Data Processing**
